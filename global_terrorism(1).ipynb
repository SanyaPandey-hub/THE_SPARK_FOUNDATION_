# COUNTRIES WITH HIGHEST TERROR ATTACKS

```python
col = ['Year','Country','Groupname','TypeofAttack','Target_type','Killed','Injured']
data_proc = data[~data.Killed.isna()][col].reset_index(drop=True)
data_proc.head()
data_proc.groupby(['Country'])['Killed'].sum().sort_values(ascending=False).head(10)
```

# MOST EFFECTED COUNTRIES

```python
sns.barplot(data_proc['Country'].value_counts()[:10].index,data_proc['Country'].value_counts()[:10].values)
plt.xlabel('Countries')
plt.ylabel('Number of attacks')
plt.xticks(rotation= 90)
plt.show()
```

# HOTZONE REGION

```python
def color_from_int(num,max_value):
    red = num/max_value
    blue = 1.0 - (num/max_value)
    green = 1.0 - (num/max_value)
    return (red, green, blue)
data_temp = data.groupby(["Country"])["Id"].count().reset_index(name="attack_count").sort_values("attack_count",ascending=False)
min_val = data_temp.loc[0:15]["attack_count"].min()
max_val = data_temp.loc[0:15]["attack_count"].max()
plt.figure(figsize=(10,10))
plt.title("Countries with highest attack counts")
plt.bar(x=data_temp[0:15]["Country"],height=data_temp[0:15]["attack_count"],align='center',color=data_temp.loc[0:15]["attack_count"].apply(lambda x, max_val=max_val: color_from_int(x,max_val)))
plt.xticks(rotation='vertical')
plt.show()
```

# Number Of Terrorist Activities Each Year

```python
print("We have data for Year {0} to {1}".format(data["Year"].min(),data["Year"].max()))
data_temp = data.groupby("Year")["Id"].count().reset_index(name="attacks")#.sort_values("attacks",ascending=False)
#max_val = data_temp["attacks"].max()
#display(data_temp)
plt.figure(figsize=(10,10))
plt.title("Year wise attack counts")
plt.xlabel("Years")
plt.ylabel("Number of attacks")
plt.plot(data_temp["Year"],data_temp["attacks"])
#plt.bar(x=data_temp["Period"],height=data_temp["attacks"],align='center',color=data_temp["attacks"].apply(lambda x,max_val=max_val : color_from_int(x,max_val)))
#plt.xticks(rotation='vertical')
plt.show()
```

```python
#Attacks Distribution Of Regions By Year
data_region=pd.crosstab(data.Year,data.region)
data_region.plot(color=sns.color_palette('Set1',12))
fig=plt.gcf()
fig.set_size_inches(20,15)
plt.show()
```
